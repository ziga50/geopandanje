```python
import geopandas as gpd
```

```python
ob = gpd.read_file(r"C:\Users\zigao\OneDrive - Univerza v Ljubljani\Documents\Nova mapa\Gurs_Arso_gikl\OB\OB.shp")
out_dir = r"C:\Users\zigao\OneDrive - Univerza v Ljubljani\Documents\Nova mapa\Gurs_Arso_gikl\OB\obcine_posamezno"
```

```python
for index, row in ob.iterrows():
    # Create a GeoDataFrame with the current row
    gdf = gpd.GeoDataFrame([row], geometry="geometry", crs=ob.crs)
    
    # Extract the value from the "ID" column for the filename
    filename = row["OB_UIME"]
    
    # Construct the output file path for the shapefile
    output_file = f"{out_dir}/{filename}.shp"  # Use / as directory separator
    
    # Save the GeoDataFrame to a shapefile
    gdf.to_file(output_file)
```
